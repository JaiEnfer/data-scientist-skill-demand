### Step 2: Data Cleaning

```python
import re

def clean_job_text(text: str)->str:
    if pd.isna(text):
        return ""
    text = str(text)

    # normalize common unicode / whitespaces
    text = text.replace("\u00a0", " ")  #non breaking space
    text = re.sub(r"/s+", " ", text) #collapse whitespace
    text = text.strip().lower()


    text = re.sub(r"[^a-z0-9\+\#\.\-\/\s]", " ", text)

    #Normalise some common variants
    text = text.replace("scikit learn","scikit-learn")
    text = text.replace("machine-learning", "machine learning")
    text = text.replace("deep-learning", "deep learning")

    #collapse whitespace again after replacement
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
# Apply cleaning +quick quality check
ds_df['job_desc_clean'] = ds_df.job_desc.apply(clean_job_text)
```

```python
### Step 3: Extracting skills from job description
import re
import pandas as pd
from collections import Counter

SKILL_PATTERNS = {
    # Programming
    "python": r"\bpython\b",
    "sql": r"\bsql\b",
    "r": r"(?<!\w)r(?!\w)",  
    "java": r"\bjava\b",
    "scala": r"\bscala\b",
    "c++": r"\bc\+\+\b",
    "c#": r"\bc\#\b",

    # Data & analysis
    "statistics": r"\bstatistic(s|al)?\b",
    "experiment design": r"\bexperiment(al)? design\b|\bdesign of experiments\b|\bdoe\b",
    "a/b testing": r"\ba\/b\b|\bab testing\b|\ba b testing\b|\bexperimentation\b",
    "data visualization": r"\bdata visualization\b|\bdata visualisation\b|\bvisualization\b|\bvisualisation\b",
    "tableau": r"\btableau\b",
    "power bi": r"\bpower\s?bi\b",

    # ML / AI
    "machine learning": r"\bmachine learning\b|\bml\b",
    "deep learning": r"\bdeep learning\b|\bdl\b",
    "nlp": r"\bnlp\b|\bnatural language processing\b",
    "time series": r"\btime series\b|\bforecast(ing)?\b",
    "recommendation systems": r"\brecommend(er|ation) systems?\b",

    # Libraries
    "pandas": r"\bpandas\b",
    "numpy": r"\bnumpy\b",
    "scikit-learn": r"\bscikit\-learn\b|\bsklearn\b|\bscikit learn\b",
    "tensorflow": r"\btensorflow\b|\btf\b",
    "pytorch": r"\bpytorch\b|\btorch\b",
    "xgboost": r"\bxgboost\b",
    "lightgbm": r"\blightgbm\b|\blgbm\b",

    # Big data
    "spark": r"\bspark\b|\bpyspark\b",
    "hadoop": r"\bhadoop\b",
    "kafka": r"\bkafka\b",

    # Cloud / DevOps
    "aws": r"\baws\b|\bamazon web services\b",
    "gcp": r"\bgcp\b|\bgoogle cloud\b|\bgoogle cloud platform\b",
    "azure": r"\bazure\b",
    "docker": r"\bdocker\b",
    "kubernetes": r"\bkubernetes\b|\bk8s\b",
    "git": r"\bgit\b|\bgithub\b|\bgitlab\b",
    "linux": r"\blinux\b",

    # Data engineering / DB
    "etl": r"\betl\b|\bextract transform load\b",
    "airflow": r"\bairflow\b",
    "dbt": r"\bdbt\b",
    "snowflake": r"\bsnowflake\b",
    "bigquery": r"\bbigquery\b",
}
```

```python
def extract_skills(text:str, patterns:dict)->str:
    if pd.isna(text) or not isinstance(text, str):
        return []
    found = []
    for skills, pat in patterns.items():
        if re.search(pat, text):
            found.append(skills)
    return found
            
```

```python
ds_df["skills"] = ds_df["job_desc_clean"].apply(lambda x: extract_skills(x, SKILL_PATTERNS))
ds_df[["job_title", "company_name", "skills"]].head(10)
```

```python
### Step 4: Coomputing skills demand(frequency)
n_jobs = len(ds_df)
skill_counter = Counter([s for skills in ds_df["skills"] for s in skills])

skill_demand = (
    pd.DataFrame(skill_counter.items(), columns=["skill", "count"]).sort_values("count", ascending= False).assign(percent=lambda d:(d["count"] / n_jobs *100).round(1))
)

skill_demand.head(10)
```

```python
### Step 5: Lets plot top 20 skills
import matplotlib.pyplot as plt

top20 = skill_demand.head(20).sort_values("count")

plt.figure(figsize = (10, 7))
plt.barh(top20["skill"], top20["count"])
plt.title("Top 20 Skills in Data Scientist Job Ads (n={})".format(n_jobs))
plt.xlabel("Number of job ads mentioning skills")
plt.ylabel("Skill")
plt.show()
```

```python
### Step 6: Skill Co-Occurrence Analysis
from itertools import combinations
from collections import Counter

pair_counter = Counter()

for skills in ds_df["skills"]:
    for pair in combinations(sorted(skills),2):
        pair_counter[pair] += 1
pair_df = (
    pd.DataFrame(pair_counter.items(), columns=["pair","count"]).sort_values("count", ascending = False)
)

pair_df.head(10)
```

```python
top_skills = skill_demand.head(15)["skill"].tolist()

matrix = pd.DataFrame(
    0, index=top_skills, columns=top_skills
)

for (s1, s2), count in pair_counter.items():
    if s1 in top_skills and s2 in top_skills:
        matrix.loc[s1, s2] = count
        matrix.loc[s2, s1] = count
```

```python
import seaborn as sns

plt.figure(figsize=(10,8))
sns.heatmap(matrix, annot=True, fmt='d', cmap= 'Blues')
plt.title("Skill Co-occurence in Data Scientist jobs")
plt.show()


```
